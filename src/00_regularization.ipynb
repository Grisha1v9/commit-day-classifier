````python
def crossval(estimator, X, y, n_splits=10): #estimator - модель 
    train_scores = []
    valid_scores = [] #обобщение, рез-т работы на новых данных
    cv = StratifiedKFold(n_splits=n_splits) #кросс-валидатор, каждый fold - n_splits разбивает выборку train на кол-во n
    
    for train, valid in cv.split(X, y): #cv.split(X, y) генератор пар массивов (train, valid) при n_splits = 10 мы передаем по факту X_train, y_train 
        X_tr = X.iloc[train]   # Берёт строки 135, 136, ..., 1350 из X    это все для 1-й иттерации
        X_val = X.iloc[valid]  # Берёт строки 0, 1, ..., 134 из X
        y_tr = y.iloc[train]   # Берёт строки 135, 136, ..., 1350 из y
        y_val = y.iloc[valid]  # Берёт строки 0, 1, ..., 134 из y

        estimator.fit(X_tr, y_tr)
        y_train_pred = estimator.predict(X_tr)
        y_valid_pred = estimator.predict(X_val)
        
        train_scores.append(accuracy_score(y_tr, y_train_pred))
        valid_scores.append(accuracy_score(y_val, y_valid_pred))
    
    for train, valid in zip(train_scores, valid_scores): #списки в пары
        print(f'train -  {train:.5f}   |   valid -  {valid:.5f}')
    
    print(f'Average accuracy on crossval is {np.mean(valid_scores):.5f}')
    print(f'Std is {np.std(valid_scores):.5f}')

logreg = LogisticRegression(fit_intercept=False, random_state=21)
crossval(logreg, X_train, y_train)
````

````python
logreg_none = LogisticRegression( 
    penalty='none', #миним пенальти
    solver='lbfgs',      # работает с 'none' и 'l2'
    fit_intercept=False,
    random_state=21,
    max_iter=1000
)
score_none = crossval(logreg_none, X_train, y_train)
````

````python
logreg_l2 = LogisticRegression( #С - параметр регуляризации
    penalty='l2', #сжатие + такие настройки стоят по дефолту 
    C=1,
    solver='lbfgs',      # работает с 'l2'
    fit_intercept=False,
    random_state=21,
    max_iter=1000
)
score_l2 = crossval(logreg_l2, X_train, y_train)
# Штраф L2 = w₁² + w₂² + w₃² + ... + wₙ²
# w = [2, -3, 1]
# L2 штраф = 2² + (-3)² + 1² = 4 + 9 + 1 = 14
````

````python
logreg_l1 = LogisticRegression(
    penalty='l1', #вес обнуляется и выкидывается из модели
    C=1,
    solver='liblinear',  # работает с 'l1' и 'l2'
    fit_intercept=False,
    random_state=21,
    max_iter=1000
)
score_l1 = crossval(logreg_l1, X_train, y_train)
# Штраф L1 = |w₁| + |w₂| + |w₃| + ... + |wₙ|
#w = [2, -3, 1]
#L1 штраф = |2| + |-3| + |1| = 2 + 3 + 1 = 6
````

## 3. SVM regularization

### a. Default regularization

1. Train a baseline model with the only parameters `probability=True`, `kernel='linear'`, `random_state=21`.
2. Use stratified K-fold cross-validation with `10` splits to evaluate the accuracy of the model.
3. The format of the result of the code where you trained and evaluated the baseline model should be similar to what you have got for the logreg.

````python
%%time
svm_baseline = SVC(
    probability=True,
    kernel='linear',
    random_state=21,
)
svm_score = crossval(svm_baseline, X_train, y_train)
````

### b. Optimizing regularization parameters

1. In the cells below try different values of the parameter `C`.

````python
C_values = [0.1, 1.0, 10.0]
svm_results = []

for C in C_values:
    print(f"C = {C}")
    
    svm_model = SVC(
        probability=True,
        kernel='linear',
        C=C,
        random_state=21
    )

    score = crossval(svm_model, X_train, y_train)
    svm_results.append({'C': C, 'accuracy': score})
````

## 6. Predictions

1. Choose the best model and use it to make predictions for the test dataset.
2. Calculate the final accuracy.
3. Analyze: for which weekday your model makes the most errors (in % of the total number of samples of that class in your test dataset).
4. Save the model.

````python
from sklearn.model_selection import GridSearchCV
import joblib
````

````python
forest = RandomForestClassifier(n_estimators=50, random_state=21)
````

````python
param_grid = {
    'criterion': ['gini', 'entropy'], #критерии качества
    'max_depth': np.arange(3, 21),
    'min_samples_split': [2, 3, 4] #разделение узлов
}
````

````python
cv_forest = GridSearchCV( #кросс-валид
    forest, 
    param_grid, 
    scoring='accuracy', #метрика
    n_jobs=-1, #использовать все ядра процесса для более быстрого рез-та
    cv=10 #кол-во фолдов - 10 иттераций
)
````

````python
cv_forest.fit(X_train, y_train)
````

````python
cv_forest.best_params_
````

````python
cv_forest.best_score_
````

````python
best_mod = cv_forest.best_estimator_
y_pred = best_mod.predict(X_test)
accuracy_score(y_test, y_pred)
````

````python
results = pd.DataFrame({
    'true': y_test.values,
    'pred': y_pred,
    'correct': y_test.values == y_pred
})

day_names = {0: 'Пн', 1: 'Вт', 2: 'Ср', 3: 'Чт', 4: 'Пт', 5: 'Сб', 6: 'Вс'}

error_rates = {}
for day in sorted(y_test.unique()):
    day_data = results[results['true'] == day]
    total = len(day_data)
    errors = (~day_data['correct']).sum()
    error_rate = errors / total * 100
    error_rates[day] = error_rate
    day_name = day_names.get(day, 0)
    print(f"{day_name}: {error_rate:.2f}% ошибок ({errors} из {total})")

max_error_day = max(error_rates, key=error_rates.get) # значение из словаря для каждого ключа -> вывод ключ
max_error_rate = error_rates[max_error_day]
max_error_name = day_names.get(max_error_day, 0)

print(f"наиб % ошибок: {max_error_name} (День {max_error_day})")
print(f"процент ошибок: {max_error_rate:.2f}%")
````
